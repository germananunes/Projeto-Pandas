### Exercício

A <a href="https://salaries.ai-jobs.net/download/">Ai-jobs</a> coleta informações salariais anonimamente de profissionais de todo o mundo no espaço de IA/ML e Big Data e as disponibiliza publicamente para qualquer pessoa usar, compartilhar e brincar. Os dados estão sendo atualizados regularmente com novos dados chegando, geralmente semanalmente.

O objetivo principal é ter dados que possam fornecer uma melhor orientação em relação ao que está sendo pago globalmente. Assim, iniciantes, profissionais experientes, gerentes de contratação, recrutadores e também fundadores de startups ou pessoas que desejam fazer uma mudança de carreira podem tomar decisões mais bem informadas.

O conjunto de dados contêm as seguintes colunas, seguidas de uma breve descrição:

- **work_year**: O ano em que o salário foi pago.
- **experience_level**: O nível de experiência do profissional durante o ano (possíveis valoeres):
        EN: Entry-level / Junior
        MI: Mid-level / Intermediate
        SE: Senior-level / Expert
        EX: Executive-level / Director
- **employment_type**: Tipo de carga horária de trabalho (valores possíveis):
        PT: Part-time
        FT: Full-time
        CT: Contract
        FL: Freelance
- **job_title**: Função de trabalho durante o ano.
- **salary**: O valor total do salário bruto pago.
- **salary_currency**: A moeda do salário pago.
- **employee_residence**: Endereço do empregado.
- **remote_ratio**: A quantidade total de trabalho feito remotamente, os valores possíveis são os seguintes:
        0: Presencial (menos de 20%)
        50: Híbrido
        100: Full Remoto (mais de 80%)
- **company_location**: O país em que o empregado trabalha ou foi contratado.
- **company_size**: O número médio de pessoas que trabalharam para a empresa durante o ano:
        S: menos de 50 empregados (pequena)
        M: 50 a 250 empregados (média)
        L: mmais de 250 empregados (grande)

5. Os funcionários com o cargo de `Data Scientist` que trabalham full remoto são mais bem pagos em média, do que os `Data Scientists` que trabalham presencialmente?
<br><br>
Responda esta pergunta agrupando estes funcionários por: <b>work_year</b>, <b>experience_level</b> e <b>salary_currency</b>

```python
salaries[salaries['job_title'] == 'Data Scientist'].groupby(
    [
    'remote_ratio',
    'job_title', 
    'salary_currency', 
    'experience_level'
    ]
).mean().reset_index()[
    [
        'job_title',
        'work_year',
        'experience_level',
        'salary_currency'
        ]
        ]


```

```python
#outra forma que eu acho que pode ser melhor para comparar, tendo em vista que todos os salarios estão em dolar

salaries[salaries['job_title'] == 'Data Scientist'].groupby(
    [
    'remote_ratio',
    'job_title', 
    'salary_in_usd', 
    'experience_level'
    ]
).mean().reset_index()[
    [
        'remote_ratio',
        'job_title',
        'work_year',
        'experience_level',
        'salary_in_usd'
        ]
        ]
```
